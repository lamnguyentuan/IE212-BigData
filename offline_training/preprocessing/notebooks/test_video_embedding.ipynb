```python
from pathlib import Path
from offline_training.preprocessing.video import (
    VideoFrameExtractor,
    VideoFrameLoader,
    TimeSformerVideoEncoder
)

# 1. Extract frames (bronze → silver)
vf = VideoFrameExtractor(
    base_dir=Path("tiktok-data"),
    num_frames=16,
    frame_size=(224, 224),
)
vf.extract_all()

# 2. Load frames
frames_dir = Path("tiktok-data/silver/123456789/frames")
loader = VideoFrameLoader()
frames = loader.load_frames(frames_dir)

# 3. Encode video embedding
encoder = TimeSformerVideoEncoder()
embedding = encoder.encode(frames)

print(embedding.shape)  # (768,) hoặc tuỳ model
```
